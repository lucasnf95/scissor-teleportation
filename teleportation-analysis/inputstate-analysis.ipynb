```python
input_folder = r"C:\\Users\\lunfa\\Documents\\PhD\\Python\\teleportation\\files\\input state\\"

dates = []
for file in os.listdir(input_folder):
    dates.append(file + '\\')

voltages = [None for date in dates]
extra = [None for date in dates]
folders = [None for date in dates]
input_files = [None for date in dates]
for d, date in enumerate(dates):
    files = os.listdir(os.path.join(input_folder + date))
    voltages[d] = [file + '\\' for file in files if '.txt' not in file]
    
for d in range(len(dates)):
    extra[d] = ['' for voltage in voltages[d]]
    folders[d] = ['' for voltage in voltages[d]]
    input_files[d] = [None for voltage in voltages[d]]
for d in range(len(dates)):
    for v, voltage in enumerate(voltages[d]):
        files = os.listdir(os.path.join(input_folder + dates[d] + voltage))
        extra[d][v] = [file + '\\' for file in files if '.trc' not in file]
        if extra[d][v] == []:
            extra[d][v] = ''

for d in range(len(dates)):
    for v in range(len(voltages[d])):
        folders[d][v] = ['' for e in extra[d][v]]
        input_files[d][v] = [None for e in extra[d][v]]
for d in range(len(dates)):
    for v in range(len(voltages[d])):
        if len(extra[d][v]) == 0:
            folders[d][v].append(os.path.join(input_folder + dates[d] + voltages[d][v] + extra[d][v]))
            input_files[d][v].append(None)
        for e in range(len(extra[d][v])):
            folders[d][v][e] = os.path.join(input_folder + dates[d] + voltages[d][v] + extra[d][v][e]) 
```

```python
for d in range(len(folders)):
    for v in range(len(folders[d])):
        for e in range(len(folders[d][v])):
            input_files[d][v][e] = inputstate.inputstate(folders[d][v][e], dimension, mf, input_plot = False)
```

```python
for d, date in enumerate(dates):
    for v in range(len(input_files[d])):
        for f, fv in enumerate(input_files[d][v]):
            print(date, fv.clearance)
```

```python
possible_voltages = ['001', '002', '003', '004', '005']

files_voltage = [[] for v in possible_voltages]
files_alpha = [[] for v in possible_voltages]
files_theta = [[] for v in possible_voltages]
amount = np.zeros(len(possible_voltages))
day = [[] for v in possible_voltages]
for d, date in enumerate(dates):
    # Ignoring data with bad clearance
    if date == '20240312\\':
        pass
    else:
        for v, voltage in enumerate(voltages[d]):
            for p, pv in enumerate(possible_voltages):
                if pv in voltage:
                    files_voltage[p].append(input_files[d][v])
                    for f in range(len(input_files[d][v])):
                        files_alpha[p].append(input_files[d][v][f].input_alpha)
                        files_theta[p].append(input_files[d][v][f].input_angle)
                        amount[p] += 1
                        day[p].append(date)
```

```python
size = (10, 8)
for p, pv in enumerate(possible_voltages):
    x_ticks = day[p]
    
    plt.figure(figsize = size)
    plt.grid()
    plt.plot(np.arange(amount[p]), files_alpha[p], 'o', color = 'orange')
    plt.xticks(np.arange(amount[p]), day[p])
    plt.xlabel('Date')
    plt.ylabel('|\\alpha|')
    plt.title('Amplitude of input state for voltage ' + pv)
    
    plt.figure(figsize = size)
    plt.grid()
    plt.plot(np.arange(amount[p]), files_theta[p], 'o', color = 'm')
    plt.xticks(np.arange(amount[p]), day[p])
    plt.xlabel('Date')
    plt.ylabel('\\theta')
    plt.title('Angle of input state for voltage ' + pv)
```

```python
alpha_day = {date: [] for date in dates}
dict_alpha_std = {date: [] for date in dates}
theta_day = {date: [] for date in dates}
dict_theta_std = {date: [] for date in dates}
for d, date in enumerate(dates):
    for v, volt in enumerate(voltages[d]):
        alphas = [None for f in input_files[d][v]]
        thetas = [None for f in input_files[d][v]]
        for f, fv in enumerate(input_files[d][v]):
            alphas[f] = fv.input_alpha
            thetas[f] = fv.input_angle
        mean_alpha = np.mean(alphas)
        std_alpha = np.std(alphas)
        mean_theta = np.mean(thetas)
        std_theta = np.std(thetas)
        alpha_day[date].append(mean_alpha)
        theta_day[date].append(mean_theta)
        dict_alpha_std[date].append(std_alpha)
        dict_theta_std[date].append(std_theta)
```

```python
for d, date in enumerate(dates):
    x_ticks = voltages[d]
    
    plt.figure(figsize = size)
    plt.errorbar(np.arange(len(voltages[d])), alpha_day[date], yerr = dict_alpha_std[date], fmt = 'o', color = 'orange')
    plt.xticks(np.arange(len(voltages[d])), voltages[d])
    plt.grid()
    plt.xlabel('Voltage')
    plt.ylabel('|\\alpha|')
    plt.title('Input state calibration on day ' + date)
    
    plt.figure(figsize = size)
    plt.errorbar(np.arange(len(voltages[d])), theta_day[date], yerr = dict_theta_std[date], fmt = 'o', color = 'm')
    plt.xticks(np.arange(len(voltages[d])), voltages[d])
    plt.grid()
    plt.xlabel('Voltage')
    plt.ylabel('\\theta')
    plt.title('Angle of input state on day ' + date)
```
