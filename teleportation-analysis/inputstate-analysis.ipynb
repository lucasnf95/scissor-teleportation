```python
input_folder = r"C:\\Users\\qpitlab\\Desktop\\Scissor QKD data\\teleportation\\input state\\"

dates = []
for file in os.listdir(input_folder):
    dates.append(file + '\\')

voltages = [None for date in dates]
for n, date in enumerate(dates):
    files = os.listdir(os.path.join(input_folder + date))
    voltages[n] = [file + '\\' for file in files if '.txt' not in file]

extra = [['' for voltage in voltages] for date in dates]
for n in range(len(voltages)):
    for v, voltage in enumerate(voltages[n]):
        files = os.listdir(os.path.join(input_folder + dates[n] + voltage))
        extra[n][v] = [file + '\\' for file in files if '.trc' not in file]
        
folders = [[['' for e in extra] for voltage in voltages] for date in dates]
for d in range(len(dates)):
    for v in range(len(voltages[d])):
        for e in range(len(extra[d][v])):
            folders[d][v][e] = os.path.join(input_folder + dates[d] + voltages[d][v] + extra[d][v][e])
```
